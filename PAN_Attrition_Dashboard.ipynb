# Palo Alto Networks — Workforce Attrition Dashboard
Run each cell in order. Steps: **Install → Upload CSV → Write app → Launch**

```python
# Cell 1: Install
import subprocess, sys
subprocess.check_call([sys.executable,'-m','pip','install','--quiet',
    'streamlit','plotly','pyngrok','pandas','numpy'])
print('All libraries installed!')
```

```python
# Cell 2: Upload CSV
from google.colab import files
import shutil
uploaded = files.upload()
for name in uploaded:
    shutil.copy(name, 'Palo_Alto_Networks__1_.csv')
    print('Saved as: Palo_Alto_Networks__1_.csv')
```

```python
# Cell 3 — Write the Streamlit app to disk (base64 safe)
import base64, pathlib

APP_B64 = ''

app_code = base64.b64decode(APP_B64.encode()).decode()
pathlib.Path('app.py').write_text(app_code)
print('ok app.py written, lines:', len(app_code.splitlines()))
```

```python
# Cell 4: Launch dashboard
NGROK_TOKEN = ''  # <- PASTE YOUR TOKEN HERE

import subprocess, threading, time
from pyngrok import ngrok, conf

conf.get_default().auth_token = NGROK_TOKEN

def run_st():
    subprocess.run(['streamlit','run','app.py',
        '--server.port','8501',
        '--server.headless','true',
        '--server.enableCORS','false'])

t = threading.Thread(target=run_st, daemon=True)
t.start()
time.sleep(6)

tunnel = ngrok.connect(8501)
url = tunnel.public_url
print('=' * 60)
print('Dashboard is LIVE!')
print('Open:', url)
print('=' * 60)
print('Keep this cell running. Stop it to shut down.')

# Keep alive
while True:
    time.sleep(30)
```
